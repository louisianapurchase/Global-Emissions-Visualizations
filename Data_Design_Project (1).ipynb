```python
# Clear Line Graph

import altair as alt
import pandas as pd
data = pd.read_csv("https://drive.google.com/uc?id=1veeMj8U7yRNUbz7EBG0p2oPJ91fwcFzJ")
alt.renderers.enable('default')

data_line = data[['Country', 'Year', 'Value']]
selected_countries = ['European Union – 27 countries (from 01/02/2020)', 'United States', 'Russia', "China (People's Republic of)", 'Japan','OECD Asia Oceania']
data_line = data[data['Country'].isin(selected_countries)]
data_line = data[(data['Country'].isin(selected_countries)) & (data['Year'] <= 2014)].copy()

data_line['Year'] = data_line['Year'].astype(int)


# Create a line chart using the filtered data
data_line['LegendLabel'] = data_line['Country'].replace({
    'European Union – 27 countries (from 01/02/2020)': 'EU',
    "China (People's Republic of)": 'China'
})

# Create a line chart using the filtered data
chart = alt.Chart(data_line).mark_line().encode(
    x = alt.X('Year:O', title='Year (1990 - 2014)'),
    y = alt.Y('Value', title='Emissions (Tonnes of CO2 equivalent)'),
    color = alt.Color('LegendLabel:N', title = 'Countries'),
    tooltip=['Country', 'Year', 'Value']
).properties(
    title = {'text': 'Emissions Over Time by Country (1990 - 2014)', 'offset': 15}
).configure_axis(
    titlePadding = 15
)
chart
```

```python
# Manipulative Line Graph

import altair as alt
import pandas as pd
data = pd.read_csv("https://drive.google.com/uc?id=1veeMj8U7yRNUbz7EBG0p2oPJ91fwcFzJ")


data_line = data[['Country', 'Year', 'Value']]
selected_countries = ['European Union – 27 countries (from 01/02/2020)', 'United States', 'Russia', "China (People's Republic of)", 'Japan','OECD Asia Oceania']
data_line = data[data['Country'].isin(selected_countries)]
data_line = data[(data['Country'].isin(selected_countries)) & (data['Year'] <= 2014)].copy()

data_line['Year'] = data_line['Year'].astype(int)

# Create a line chart using the filtered data
data_line['LegendLabel'] = data_line['Country'].replace({
    'European Union – 27 countries (from 01/02/2020)': 'EU',
    "China (People's Republic of)": 'CHN',
    "United States": "USA",
    "Japan": "JPN",
    "Russia": "RUS",
    'OECD Asia Oceania': 'OECD'
})

# Define custom RGB colors for each country
color_mapping = {
    'EU': 'rgb(0, 0, 255)',
    'USA': 'rgb(0, 0, 200)',
    'RUS': 'rgb(0, 0, 150)',
    'CHN': 'rgb(0, 0, 100)',
    'JPN': 'rgb(0, 0, 50)',
    'OECD': 'rgb(0,0,0)'
}


# Create a line chart using the filtered data
chart = alt.Chart(data_line).mark_line().encode(
    x = alt.X('Year:O', title = 'Year (1990 - 2014)'),
    y = alt.Y('Value', title = 'Emissions'),
    color = alt.Color('LegendLabel:N', title='Countries', scale=alt.Scale(domain = list(color_mapping.keys()), range = list(color_mapping.values())))
).properties(
    title = {'text': 'Emissions Over Time by Country (1990 - 2014)', 'offset': 15}
).configure_axis(
    grid = False,
    titlePadding = 15
)
chart
```

```python
import altair as alt
import pandas as pd

color_scheme = 'redblue'


missing_data = pd.DataFrame(columns=['Year', 'Country'])
for year in data_line1['Year'].unique():
    for country in data_line1['Country'].unique():
        if data_line1[(data_line1['Year'] == year) & (data_line1['Country'] == country)]['Value'].isna().all():
            missing_data = missing_data.append({'Year': year, 'Country': country}, ignore_index=True)


data_line1 = pd.concat([data_line1, missing_data])

data_line1['HasData'] = data_line1['Value'].notna()

base_chart = alt.Chart(data_line1).mark_rect().encode(
    x=alt.X('Year:O', title="Year (1990 - 2014)"),
    y=alt.Y('Country:N', title="Country"),
    color=alt.Color('Value:Q', title='Total Emissions (Tonnes of CO2 equivalent)', scale=alt.Scale(scheme=color_scheme, domain=[1500000, 13000000,])),
    tooltip=[
        alt.Tooltip('Year:O', title="Year"),
        alt.Tooltip('Value:Q', title='Total Emissions (Tonnes of CO2 equivalent)'),
        alt.Tooltip('Country:N', title="Country"),
    ]
)

text_chart = alt.Chart(data_line1).mark_text(baseline='middle').encode(
    x=alt.X('Year:O'),
    y=alt.Y('Country:N'),
    text=alt.condition(
        alt.datum.HasData,
        alt.value(''),  # Use an empty string when there is data
        alt.value('X')  # Display 'X' when there is no data
    )
)

chart = (base_chart + text_chart).properties(
    width=500,
    height=300
)

chart
#######
```
